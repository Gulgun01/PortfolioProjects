```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
import numpy as np
from scipy.stats import shapiro
# load Apple stock data
apple_stock = yf.Ticker("AAPL")

#Specify date range
start_date = "2014-01-01"
end_date = "2024-08-03"

# Pull daily stock data
apple_stock_data = apple_stock.history(start=start_date, end=end_date)

# Set data to UTC and remove time zone
apple_stock_data.index = apple_stock_data.index.tz_localize(None)

# View data
print(apple_stock_data.head(10))

# Save to CSV file
apple_stock_data.to_csv("apple_stock_data.csv")

# Read CSV file as pandas DataFrame
data = pd.read_csv('apple_stock_data.csv')

#Viewlast 10 rows of  dataset 
print(data.tail(10))

















```

```python

# size of data set
data.shape

# checking null values
null_check =  data.isnull().any()
print(null_check)

# duplicate data in the dataset
data.duplicated().sum()

# Outlier detection function with IQR method
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Detect and print outliers for each column
for column in data.select_dtypes(include=[float, int]).columns:
    outliers = detect_outliers_iqr(data, column)
    print(f"Outliers in {column} using IQR method:")
    print(outliers)

# Visualize outliers with box plot
data.boxplot()
plt.show()

# By using the z score method, we will statistically show whether these 3 columns (Volume, Dividends, Stock Splits) are outliers.



# Outlier detection with Z-Score method
def detect_outliers_zscore(data, column):
    z_scores = stats.zscore(data[column])
    abs_z_scores = abs(z_scores)
    outliers = data[abs_z_scores > 3]
    return outliers

# Detect and print outliers for each column
for column in apple_stock_data.columns:
    outliers = detect_outliers_zscore(apple_stock_data, column)
    print(f"Outliers in {column} using Z-Score method:")
    print(outliers)

# Print each unique value in the 'Stock Splits' column and their count
stock_splits_counts = data['Stock Splits'].value_counts()
print(stock_splits_counts)
     
#This column will be deleted because there are only 2 unique values. 
#Remove 'Stock Splits' column
      
data = data.drop(columns=['Stock Splits'])

# print each unique value in the 'Dividends' column and their count
Dividends_counts = data['Dividends'].value_counts()

print(Dividends_counts)
# Remove 'Dividends' column
      
data = data.drop(columns=['Dividends'])


# Print each unique value in the Volume column and their count
Volume_counts = data['Volume'].value_counts()

print(Volume_counts)

```

```python

# Visualizing the distribution of data with a histogram
plt.figure(figsize=(12, 6))
sns.histplot(data['Volume'], bins=50, kde=True)
plt.title('Volume Distribution')
plt.show()

# VThe data in the measurement column show positive skewness. Therefore, the log method will be used. The log transformation can make the data more symmetric by reducing the effect of high values.

# applying log transformation
data['Volume_log'] = np.log1p(data['Volume'])

#Visualizing data after log transformation
plt.figure(figsize=(12, 6))
sns.histplot(data['Volume_log'], bins=50, kde=True)
plt.title('Log Transformed Volume Distribution')
plt.show()
# After log transformation, the data approached normal distribution very closely.
#They are still outliers therfore we will use Capping method

# determining lower and upper bounds
lower_cap = data['Volume_log'].quantile(0.05)  # Alt %5 snr
upper_cap = data['Volume_log'].quantile(0.95)  # st %95 snr
# applying Capping 
data['Volume_log_capped'] = np.clip(data['Volume_log'], lower_cap, upper_cap)


# Visualizing capped data
plt.figure(figsize=(12, 6))
sns.histplot(data['Volume_log_capped'], bins=50, kde=True)
plt.title('Capped Volume Log Distribution')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x=data['Volume_log_capped'])
plt.title('Capped Volume Log Boxplot')
plt.show()

# type of columns in the dataset
data.dtypes


# Converting date format data to datetime data type and specifying UTC
data['Date'] = pd.to_datetime(data['Date'], utc=True)

# Creating Year, Month and Day columns
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day
# View results
print(data)

# Turn off scientific notation
pd.set_option('display.float_format', '{:,.0f}'.format)






```

```python
#Null Hypothesis (H0): There is no significant difference between the average closing prices between the pre- and post-corona period.
#Alternative Hypothesis (H1): There is a significant difference between the average closing prices between the pre- and post-corona period.  
```

```python



#  load dataset
df = pd.read_csv('apple_stock_data.csv')

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce', utc=True)

# Remove UTC time zone information
df['Date'] = df['Date'].dt.tz_localize(None)

#  Define relevant periods 
pre_corona_start = pd.to_datetime('2015-10-18')
pre_corona_end = pd.to_datetime('2020-03-10')
post_corona_start = pd.to_datetime('2020-03-11')
post_corona_end = pd.to_datetime('2024-08-02')


# Filter data by period
pre_corona = df[(df['Date'] >= pre_corona_start) & (df['Date'] <= pre_corona_end)]
post_corona = df[(df['Date'] >= post_corona_start) & (df['Date'] <= post_corona_end)]

# Just take the 'Close' column
pre_corona_close = pre_corona[['Date', 'Close']]
post_corona_close = post_corona[['Date', 'Close']]




print(pre_corona_close)
print(post_corona_close)
```

```python

# Round prices to two decimal places
df['Close'] = df['Close'].round(2)

# Set float format
pd.options.display.float_format = '{:.2f}'.format

#Check the last 10 rows and the 'Close' column
print(df[['Date', 'Close']].tail(10))

# Find the number of null values
null_counts = df.isnull().sum()

# Print the number of null values in each column
print(null_counts)



```

```python
#Two sample t-test will be applied. Normality and variance homogeneity will be tested first.
#normality assumption:
# Sapiro-Wilk test

# # Get only 'Close' columns of pre-corona and post-corona periods
pre_corona_close_values = pre_corona_close['Close']
post_corona_close_values = post_corona_close['Close']

# Perform the Sapiro-Wilk test
stat_pre, p_pre = shapiro(pre_corona_close_values)
stat_post, p_post = shapiro(post_corona_close_values)

# print results
print(f"Shapiro-Wilk test result in the pre-corona period: Stat={stat_pre:.4f}, p={p_pre:.4f}")
print(f"Shapiro-Wilk test result in the post-corona period: Stat={stat_post:.4f}, p={p_post:.4f}")
#Since the p value is not  less than 0.05, the null hypothesis was not rejected and the Mann-Whitney U test will be used since the normality assumption is not met.
#Mann-Whitney U test
from scipy.stats import mannwhitneyu
stat, p_value = mannwhitneyu(pre_corona_close_values, post_corona_close_values, alternative='two-sided')

print(f"Mann-Whitney U  test result: Stat={stat:.4f}, p={p_value:.4f}")

# Interpretation according to the significance of the test
alpha = 0.05
if p_value < alpha:
    print("There is a significant difference between the average close prices of the two periods.")
else:
    print("There is not a significant difference between the average close prices of the two periods.")
```

```python
#Conclusion
#"There is a significant difference between the average closing prices between the two periods. With 95% confidence, we can say that the COVID-19 pandemic has a significant impact on Apple stock prices. The null hypothesis is rejected."
```
